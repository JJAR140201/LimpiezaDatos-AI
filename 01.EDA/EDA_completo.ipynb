### Paso 1: Cargamos las librerias a utilizar

```python
# Libreria para operaciones estadísticas y aritméticas
import numpy as np
# Libreria para el manejo de datos
import pandas as pd
# Libreria para gráficas 2D
import matplotlib.pyplot as plt
```

### Paso 2: Cargamos los datos a un DataFrame

```python
# Cargamos los datos
data = pd.read_csv("EDA.csv")
data
```

### Paso 3: Exploramos los datos

```python
# Se cuentas las filas y las columnas
print("Filas  Columnas")
data.shape
```

```python
# Se identifican columnas NaN
print("Columna    Cantidad NaN")
data.isnull().sum()
```

```python
# Se eliminan las NaN del dataset
data = data.dropna()
```

```python
# Se vuelve a verificar que no existan NaN
print("Columna  Cantidad NaN")
data.isnull().sum()
```

```python
# Se cuentan los registros resultantes
print("Filas  Columnas")
data.shape
```

```python
# Se observan las estadísticas
data.describe()
```

```python
# Se filtran los datos de España
espania = data[ data['country'] == 'Spain' ]
espania
```

```python
# Se genera una gráfica de barras con los datos de España
espania.plot(kind="bar")
```

```python
# Se compara el crecimiento de la población de España y Argentina
argentina = data[ data['country'] == 'Argentina' ]

# Eje x (años)
anios = espania['year'].unique()

# Se crea un dataframe con los datos del: españa y argentina y años
data_poblacion = pd.DataFrame({'España': espania['population'].values, 'Argentina': argentina['population'].values}, index= anios )

# Se crea la gráfica comparativa
data_poblacion.plot(kind="bar")
```
